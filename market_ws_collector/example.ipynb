```python
# ✅ 示例 cell 1
import pandas as pd
import numpy as np

# 1. 导入库
import plotly.express as px
import plotly.io as pio

# 激活交互渲染（Jupyter 环境）
pio.renderers.default = "notebook"  # 如果无效，试试 "notebook_connected" 或 "iframe_connected"

# ✅ 示例 cell 2
df = pd.read_csv('./snapshots/bingx.csv')
df.head()
```

```python

# 3. 转换时间列为 datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 4. 计算差价
df['spread'] = df['ask'] - df['bid']

# 5. 显示前几条数据确认
# print(df.head())

# 6. 画差价时间序列图
fig = px.line(
    df,
    x='timestamp',
    y='spread',
    color='symbol',
    title='Symbol Spread Over Time',
    labels={'spread': 'Spread (ask - bid)', 'timestamp': 'Time'}
)

fig.update_layout(
    xaxis=dict(rangeslider=dict(visible=True)),
    hovermode='x unified'
)

fig.show()
```
